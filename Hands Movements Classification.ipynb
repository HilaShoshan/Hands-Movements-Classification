```python
def make_files(filename):
    df = pd.read_csv(filename)    
    df = df.ix[700:]
    c = re.split("[0-9,.]+",os.path.basename(filename))
    df.insert(loc=0, column='Name', value=c[0])
    df.insert(loc=1, column='Status', value=c[1])
    return df
```
